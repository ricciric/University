```python
import random as rand
'''Consideriamo tre porte chiuse, numerate con 1, 2 e 3. Dietro una delle tre porte
c’è un’automobile. Dietro ciascuna delle altre due c’è una capra.
A questo punto il gestore del gioco, che conosce dove sta l’automobile e dove
stanno le capre, ci propone di scegliere una delle tre porte.
Se si sceglie la porta con l’automobile si vince l’automobile, altrimenti si vince
una capra. L’obiettivo del gioco è ovviamente quello di indovinare la porta con
l’automobile.
Dopo la nostra scelta della porta, il gestore apre una delle due porte rimanenti
dietro la quale c’è una capra e ci chiede se vogliamo cambiare la nostra scelta
della porta o se vogliamo confermare la scelta iniziale.
Scrivere un programma Python che simuli il suddetto gioco su 10.000 prove e
che presenti la percentuale di vincita dell’automobile, sia nel caso in cui si
scelga la strategia di cambiare ogni volta la porta, sia nel caso in cui non si
cambi la scelta iniziale della porta. '''

porta = {}
porta[1] = 'automobile'
porta[2] = 'capra'
porta[3] = 'capra'


def gioco(scelta ,cambio):
        if cambio == 0:
            print('porta aperta ' + porta[scelta])
        else:
            temp = scelta
            while scelta == temp:
                scelta = rand.randint(1,3)
            print('porta aperta ' + porta[scelta])

        return porta[scelta]

def simula(n):
    volte = n
    vittorie_no_cambio = 0
    vittorie_si_cambio = 0
    while volte>0:
        if gioco(rand.randint(1,3), 0) == porta[1]:
            vittorie_no_cambio = vittorie_no_cambio + 1
        volte = volte - 1
    volte = n
    while volte>0:
        if gioco(rand.randint(1,3), 1) == porta[1]:
            vittorie_si_cambio = vittorie_si_cambio + 1
        volte = volte - 1
    print('Percentuale Vittorie con cambio: ', vittorie_si_cambio/n, 'percentuale vittorie senza cambio: ', vittorie_no_cambio/n)

simula(10000)


            
```
