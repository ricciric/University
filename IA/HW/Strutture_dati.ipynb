```python
#################################
# Liste doppiamente concatenate #
#################################

class NodeD:

    data = None
    prev = None
    next = None

    def __init__(self, data):
        self.data = data
    
    def __str__(self):
        return 'Nodo: ' + str(self.data)

class DoubleList:

    def __init__ (self):
        self.head = None
        self.tail = None
    
    def add(self, newNode):
        if(self.head == None):
            self.head = newNode
            self.tail = newNode
            newNode.next = None
            newNode.prev = None       
        else:
            newNode.prev = self.tail
            self.tail.next = newNode
            self.tail = newNode
            self.tail.next = None
    
    def pop(self, newNode):
        p = self.head
        if self.head == None:
            self.head = newNode
            self.tail = newNode
            newNode.next = None
            newNode.prev = None
        
        else:
            newNode.next = self.head
            self.head = newNode
            self.head.prev = None
    
    def delete(self, data):
        p = self.head
        if(p.data == data):
            l = p.next
            p.next = None
            del p
            l = self.head
            print('Rimosso!')
        else:
            while(p.data != data):
                l = p
                p = p.next
            l.next = p.next
            p.next.prev = l
            p.next = None
            p.prev = None
            del p
            print('Rimosso!')
    
    def __str__(self):
        stri = ''
        if self.head == None:
           return ('Lista vuota!')
        else: 
            p = self.head
            while p != None:
                stri += str(p.data) + ' ==> '
                p = p.next
            stri += 'Fine lista'
            
            return stri
    
    def search(self, data):
        p = self.head
        if(p.data == data):
            return p
        elif(self.tail.data == data):
            return self.tail
        else:
            while(p.data != data):
                p = p.next
            return p
    
lista = DoubleList()
lista.add(NodeD(20))
lista.add(NodeD(30))
lista.add(NodeD(50))
lista.add(NodeD(60))
lista.add(NodeD(70))
lista.pop(NodeD(100))

lista.delete(50)
print(lista)
```
